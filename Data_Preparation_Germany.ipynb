```python
path = '/content/drive/MyDrive/Tuebingen/Data Literacy/Project/'
data_ger=pd.read_csv(path+'btw21_wbz_ergebnisse.csv',skiprows=1,sep=';',decimal=',')
data_ger.shape
```

```python
data_ger.head(2)
```

```python
data_ger = pd.concat([data_ger.iloc[:,[0,8,9,11]], data_ger.iloc[:,65:73]],axis=1)
parties_ger=['CDU','SPD','AfD','FDP','DIE LINKE','GRUNE','CSU']
data_ger.columns = ['Wahlkreis','Gemeindename','Wahlbezirk','registered','turnout']+parties_ger
```

```python
data_ger['CDU_CSU'] = data_ger['CDU']+data_ger['CSU']
data_ger.drop(['CSU','CDU'],axis=1)
```

```python
data_ger=data_ger.loc[(data_ger['turnout']>0)&(data_ger['registered']>0)]
```

```python
pd.to_pickle(data_ger,path+'data_ger.pkl')
```
